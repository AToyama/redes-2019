## Exercício 1

Um *conjunto independente* é o oposto de clique: conjunto de nós tal que ninguém está conectado com ninguém (que triste). Se houver uma só conexão nesse grupo, não pode mais ser chamado de conjunto independente. Todo subconjunto de um conjunto independente obviamente também é independente, então encontrar "todos os conjuntos independentes" não é um problema muito interessante. Mais interessante é encontrar conjuntos independentes **maximais**, ou seja, conjuntos "no limite" tais que, se você adicionar qualquer outro nó, deixa de ser independnete.

Não há funções prontas para encontrar conjuntos independentes maximais. Nem o número deles. Mas como podemos usar o que já sabemos para fazer isso facilmente?

Tente descobrir como fazer isso na NetworkX com o menor número de linhas possível.
